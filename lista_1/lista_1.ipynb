## **Exercício 4:** 

Um analista precisa avaliar se o preço do seu fornecedor local está competitivo <br>
com a possibilidade de importação. Para isso, o analista fez uma pesquisa e viu que o valor<br>
FOB Golfo do produto estaria em US$ 500/t. O preço cobrado pelo produtor nacional é de<br>
US$ 800/t (preço entregue na planta do cliente).<br>

**Dados os valores:**

* **Frete Marítimo + seguro internacional – USD 50/t**
* **Imposto de Importação – 14% sobre o Preço CIF**
* **AFRMM – 25% sobre frete marítimo + seguro internacional**
* **Frete (Porto ao cliente) – US$ 50/t**

<br>

### **a) Apresente os conceitos de preço CIF e preço FOB.**

``FOB`` = valor do bem no porto + frete interno<br>
``CIF`` = FOB + frete internacional + seguro

### **b) O que se mostra mais vantajoso: comprar do produtor nacional ou importar?**

```python
FOB = 500
frete_mar = 50
AFRMM = frete_mar * 0.25
frete_porto = 50

CIF = FOB + frete_mar + AFRMM + frete_porto

imposto = CIF * 0.14

importar = CIF + imposto

nacional = 800

print(f'Vale mais a pena importar, já que é feita a economia de $ {nacional - importar}')
```
